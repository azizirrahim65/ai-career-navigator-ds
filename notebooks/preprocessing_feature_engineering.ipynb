```python
import pandas as pd
import re

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python

career_df = pd.read_csv("cleaned_career_recomendation.csv")

top_jobs = career_df['first_job_title'].value_counts().head(10).index
career_df = career_df[career_df['first_job_title'].isin(top_jobs)].copy()
```

```python
# Define the features and target variable
x_text = career_df['skills']
y = career_df['first_job_title']
```

```python
# Cleaning NLP
def clean_text(text):
    # Remove non-alphanumeric characters
    text = str(text).lower()
    text = re.sub(r'[^a-zA-Z ]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

career_df['clean_skills'] = career_df['skills'].apply(clean_text)
```

```python
# Vectorize the text data using TF-IDF
tfidf = TfidfVectorizer(max_features=1000)

x = tfidf.fit_transform(career_df['clean_skills'])
```

```python
# Label encode the target variable
encoder = LabelEncoder()

y = encoder.fit_transform(career_df['first_job_title'])
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42,
)
```

```python
print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
print("y_train:", y_train.shape)
print("y_test:", y_test.shape)
```

```python
import joblib

joblib.dump(tfidf, "tfidf_vectorizer.pkl")
joblib.dump(encoder, "label_encoder.pkl")
```

Tahap preprocessing dan feature engineering dilakukan untuk mengubah data teks pada kolom skills menjadi representasi numerik menggunakan TF-IDF. Target label first_job_title juga diubah menjadi angka menggunakan LabelEncoder agar dapat digunakan pada tahap modeling.
